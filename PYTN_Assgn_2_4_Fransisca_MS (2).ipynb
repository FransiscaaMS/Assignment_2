### 3. Bagaimana Probability Distribution (Uniform Distribution, Normal Distribution, Gamma Distribution, Eksponensial Distribution, Poisson Distribution) pada kolom total units?

#### **Probability Distribution**
Probability distribution adalah fungsi di bawah teori dan statistik probabilitas yang memberi kita seberapa besar kemungkinan hasil yang berbeda dalam sebuah eksperimen.<br>
Beberapa jenis probability distribution dan penggambarannya adalah sebagai berikut.

1. **Uniform Distribution** <br>
    Adalah bentuk distribusi probabilitas di mana suatu peristiwa memiliki kemungkinan yang sama untuk terjadi dalam interval tertentu.<br>
    Rumus fungsi data berdistribusi uniform sebagai berikut. <br>
    $f(x) = \begin{cases}
    \frac{1}{b-a}, \hspace{0.4cm} a \leq x \leq b \\
    0, \hspace{0.7cm} other.
    \end{cases}$
    
     dengan <br>
     $f(x)$ = Fungsi Kepadatan Probabilitas (FKP) <br>
     a = batas atas<br>
     b = batas bawah<br>
    
    **Keterangan**<br>
    Untuk Uniform Distribution, gunakan **uniform.rvs(size= jumlah data, loc= mean, scale= standard deviasi)**

```python
from scipy.stats import uniform
Tot_uniform = uniform.rvs(size=84548, loc=2, scale=19)
sns.distplot(Tot_uniform,
             bins=100)
plt.xlabel('Uniform Distribution berdasarkan total units')
plt.ylabel('Frequency')
plt.show()
```

```python
Tot_uni = uniform.rvs(size=84548, loc=2, scale=19)
graph = seaborn.displot(Tot_uni)
plt.xlabel('Uniform Distribution berdasarkan total units')
plt.ylabel('Frequency')
```

------------------------------
2. **Normal Distribution** <br> 
    Normal Distribution, juga dikenal sebagai Gaussian distribution, ada di mana-mana dalam Data Science. Kita akan menemukannya di banyak tempat terutama dalam topik inferensi statistik.

    Rumus Normal Distribution adalah sebagai berikut. <br>
    $f(x) = \frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{(x-\mu)^{2}}{2\sigma^{2}}}$. <br>
    dengan <br>
    $f(x)$ = Fungsi Kepadatan Probabilitas (FKP) <br>
    $\sigma $ = Standard deviasi <br>
    $\mu$ = Mean <br>
    $\pi$ = 3.141592... <br>
    $e$ = Natural number = 2.7182818...<br>
    $x$ = Nilai variabel

    **Keterangan**<br>
    Untuk Normal Distribution, gunakan **norm.rvs(size= jumlah data, loc= mean, scale= standard deviasi)**

```python
from scipy.stats import norm
Tot_norm = norm.rvs(size=84548, loc=2, scale=19)
plt.xlabel('Normal Distribution berdasarkan total units')
plt.ylabel('Frequency')
sns.distplot(Tot_norm)
```

```python
Tot_nor = norm.rvs(size=84548, loc=2, scale=19)
graph = seaborn.displot(Tot_nor)
plt.xlabel('Normal Distribution berdasarkan total units')
plt.ylabel('Frequency')
```

3. **Gamma Distribution** <br>
    Gamma distribution adalah two-parameter family dari continuous probability distributions. Meskipun jarang digunakan dalam bentuk mentahnya tetapi distribusi lain yang populer digunakan seperti exponential, chi-squared, erlang distributions adalah kasus khusus dari distribusi gamma.

    Rumus Gamma Distribution adalah sebagai berikut. <br>
$ f(x;\beta;\alpha) = \left\{ \begin{array}{cl}
\frac{x^{\alpha-1}}{\beta^{\alpha}\Gamma(\alpha)}e^{-\frac{x}{\beta}}  & ; x>0 \\
\ 0 & ; x~lainnya
\end{array} \right. $. <br>
    dengan <br>
    $f(x;\beta;\alpha)$ = Fungsi Kepadatan Probabilitas (FKP) <br>
    $\sigma = \beta \sqrt{\alpha} $ <br>
    $\mu = \beta \alpha $ <br>
    $\Gamma(\alpha) = (\alpha-1)!$ <br>
    $e$ = Natural number = 2.7182818...<br>
    $x$ = Nilai variabel <br>
    
    **Keterangan**<br>
    Untuk Gamma Distribution, gunakan **gamma.rvs(size= jumlah data, a = 5)** <br>
    Diambil a = 5 sebagai puncak distribusi. <br>
    Perhatikan bahwa $\Gamma(a) = (a-1)!$ merupakan fungsi gamma sehingga apabila a = 1, akan menghasilkan fungsi berdistribusi eksponensial.

```python
from scipy.stats import gamma
Tot_gam = gamma.rvs(size=84548,a=5)
plt.xlabel('Gamma Distribution berdasarkan total units')
plt.ylabel('Frequency')
sns.distplot(Tot_gam)
```

```python
Tot_gamma = gamma.rvs(size=84548,a=5)
graph = seaborn.displot(Tot_gamma)
plt.xlabel('Gamma Distribution berdasarkan total units')
plt.ylabel('Frequency')
```

4. **Exponential Distribution** <br>
Exponential distribution menggambarkan waktu antara peristiwa dalam Poisson point process, yaitu proses di mana peristiwa terjadi terus menerus dan independen pada tingkat rata-rata yang konstan.

   Rumus Exponential Distribution adalah sebagai berikut. <br>
    $ f(x;\lambda) = \left\{ \begin{array}{cl}
\lambda e^{-\lambda x}  & ; x\ge 0 \\
\ 0 & ; x \lt 0
\end{array} \right. $ <br>

    $x$ = Nilai variabel <br>
    $e$ = Natural number = 2.7182818...<br>

5. **Poisson Distribution** <br>
Poisson Distribution merupakan suatu discrete probability distribution yang memenuhi kriteria berikut:
- Random variable x merepresentasikan banyaknya kemunculan suatu event dalam interval waktu tertentu.
- Nilai probability untuk kemunculan event adalah sama untuk setiap interval.
- Jumlah kemunculan event pada suatu interval adalah independent terhadap jumlah kemunculan event pada interval lainnya.

    Rumus Poisson DistributioN adalah sebagai berikut. <br>
    $p(x) = P(X=x) = \frac{e^{-\lambda}\lambda^{x}}{x!}$ <br> 
    dengan :<br>
    X = 0,1,2... <br>
    $\mu = \lambda$ <br>
    $\sigma = \sqrt{\lambda}$
    
    **Keterangan**<br>
    Untuk Poisson Distribution, gunakan **poisson.rvs(size = jumlah data, mu = standard deviasi)**

```python
from scipy.stats import poisson
Tot_pois = poisson.rvs(size=84548,mu=360)
plt.xlabel('Poisson Distribution berdasarkan total units')
plt.ylabel('Frequency')
sns.distplot(Tot_pois)
```

### 4. Bagaimana Binomial dan Bernoulli Distribution pada kolom building class category : ONE FAMILY DWELLINGS?

1. **Binomial Distribution** <br>
    Distribusi di mana hanya dua hasil yang mungkin, seperti sukses atau gagal, untung atau rugi, menang atau kalah dan di mana probabilitas keberhasilan dan kegagalan sama untuk semua percobaan disebut Binomial Distribution. 

    Rumus peluang data berdistribusi binomial sebagai berikut. <br>
    $f(x) = \frac{n!}{(n-x)!}p^{x}q^{n-x}$ <br>
    <br>
    dengan <br>
    n = jumlah percobaan <br>
    p = peluang sukses <br>
    q = peluang gagal <br>
    n = jumlah data yang dipilih <br>
    x = total data keseluruhan

    Pada kasus ini, saya menggunakan <br>

### 5. Confidence interval (for the population Proportion). Berdasarkan data wilayah Brooklyn yang memiliki tahun dibangun 1925. Bagaimana kesimpulannya?

#### Confidence interval for the population Proportion 

Confidence interval = $\mu \pm (z_{score} * ste)$ <br>
ste = $\sqrt{\frac{p(1-p)}{n}}$. <br>

Berikut parameter-parameter yang digunakan.
- $\mu$ = rata-rata
- $z_{score}$ = nilai z
- ste = standar error for population propotion
- p = probabilitas data ($\frac{x}{n}$)
- x = jumlah data yang muncul
- n = jumlah keseluruhan data <br>

Akan ditunjukkan confidence interval berdasarkan data wilayah Brooklyn yang memiliki tahun dibangun 1925.

```python
z_score = 1.96
lcb = p - (z_score*ste)
ucb = p + (z_score*ste)
print('Diperoleh lower limit of the CI = ', lcb)
print('Diperoleh upper limit of the CI =', ucb)
```

**Kesimpulan**

Berdasarkan perhitungan diatas, diperoleh hasil sebagai berikut. <br>
Confidence interval for population Proportion = $\mu \pm (z_{score} * ste)$

- $\mu + (z_{score} * ste) = 0.06870873696885775 \approx 0.0687 $

population : orang yang membeli residential units <br>

phat : persentase 'total units' bergantung pada 'residential units' (dalam kondisi tidak bernilai 0) <br>
Null Hypothesis $(H_0)$ : pnull = phat <br>
Alternative Hypothesis $(H_1)$ : pnull < phat

Diketahui bahwa, untuk mencari phat : <br>
phat = $\frac{x}{n}$ <br>
dengan<br>
x = jumlah data yang dipilih <br>
n = total data keseluruhan

x = Jumlah residential units dan total units tidak bernilai 0 <br>

Meskipun, kita tidak menerima hipotesis alternatif, secara informal berarti bahwa ada kemungkinan besar proporsi ini lebih dari 23%.

### 9. Hypothesis Testing (for One Proportion). Pertanyaan: Jumlah proporsi penduduk Bronx yang memiliki residential units pada tahun 1901 adalah 0.35%. Apakah terdapat lebih banyak jumlah proporsi penduduk di Brooklyn yang memiliki residential units pada tahun 1915?

#### Hypothesis Testing (for One Proportion)
$\sum_1$ : Jumlah residential units di Bronx tahun 1901 <br>
$\sum_2$ : Jumlah residential units di Brooklyn tahun 1915 <br>
Null Hypothesis $(H_0) : \sum_1 = \sum_2$ <br>
Alternative Hypothesis $(H_1)$ : $\sum_1 > \sum_2$

Dalam hal ini, <br>
ster = $ \sqrt{\frac{p_0(1-p_0)}{n}} $ <br>
$p$ = $\frac{n}{x}$ <br>
dengan<br>
ster = standard error Hypothesis Testing (for One Proportion) <br>
n = jumlah data yang dipilih <br>
x = total data keseluruhan

Keterangan:<br>
Dalam kasus ini, dipilih <br>
n = jumlah proporsi penduduk Bronx yang memiliki residential units pada tahun 1901 <br>
x = total data keseluruhan <br>
$ p_0 $ = 0.35% = 0.0035

```python
OP_Bro = (df[(df['borough'] == 'Brooklyn') & (df['residential units'] != 0) & (df['year built'] == 1915)])
OP_Bro
```

```python
Brooklyn = OP_Bro.loc[0:, ['borough','residential units','year built']]
Brooklyn
```

```python
OP_Brox = (df[(df['borough'] == 'Bronx') & (df['residential units'] != 0) & (df['year built'] == 1901)])
OP_Brox
```

```python
Bronx = OP_Brox.loc[0:, ['borough','residential units','year built']]
Bronx
```

```python
Total = len(df)
Brklyn = len(Brooklyn)
Brx = len(Bronx)
print ("Jumlah residential units di Brooklyn tahun 1915: {0} ({1:2.2f}%)".format(Brklyn, (Brklyn/Total)*100))
print ("Jumlah residential units di Bronx tahun 1901: {0} ({1:2.2f}%)".format(Brx, (Brx/Total)*100))
```

Didapatkan hasil bahwa, <br>
n = jumlah proporsi penduduk Bronx yang memiliki residential units pada tahun 1901 = 295 <br>
x = total data keseluruhan = 84548 

```python
#Mari kita hitung nilai p
n = 333
x = 84548
p = n/x
print("Diperoleh nilai p = ", p)
```

```python
#Menghitung standar error 
p0 = 0.0035
ster = np.sqrt(p0 * (1-p0) / Total)
ster
```

Selanjutnya adalah melakukan test statistic dengan rumus sebagai berikut: <br>
$\frac{Best ~ estimate - hypothesized ~ estimate}{standar~ error}$ <br>
dengan<br>
Best estimate = p <br>
Hypothesized estimate = $p_0$ <br>
Standar error = sudah dihitung sebelumnya

```python
#Lakukan test statistics menggunakan rumus test statistics di atas:

#Best estimate
be = p

#hypothesized estimate
he = p0
test_stat = (be - he)/ster
test_stat
```

```python
import statsmodels.api as sm
import scipy.stats.distributions as dist

pvalue = 2*dist.norm.cdf(-np.abs(test_stat))
print(pvalue)
```

**Kesimpulan**

Pertimbangkan significance level alpha adalah 5% atau 0.05. Tingkat signifikansi 5% atau kurang berarti ada kemungkinan 95% atau lebih besar bahwa hasilnya tidak acak.

Di sini p-value lebih rendah dari significance level yang kita tentukan yaitu 0.05. Jadi, kita reject the null hypothesis. Artinya, terdapat perbedaan yang signifikan dalam proporsi penduduk yang memiliki resedential units di Bronx dan Brooklyn.

    $μ_2$ : Rata-rata commercial units di Staten Island <br>

    (0.0024920921170876347) kurang dari alpha $ (p_{value} < \alpha) $, kita dapat reject the Null hypothesis bahwa rata
    rata commercial units di Manhattan = rata-rata commercial units di Staten Island. <br>
    Catatan, kita menggunakan alternative="two-sided" dalam z-test karena di sini kita memeriksa ketidaksamaan.

9. **Hypothesis Testing (for One Proportion)** <br>
    Pertanyaan: Jumlah proporsi penduduk Bronx yang memiliki residential units pada tahun 1901 adalah 0.35%. Apakah
    terdapat lebih banyak jumlah proporsi penduduk di Brooklyn yang memiliki residential units pada tahun 1915?
    
    Diketahui:<br>
    $\sum_1$ : Jumlah residential units di Bronx tahun 1901 <br>
    $\sum_2$ : Jumlah residential units di Brooklyn tahun 1915 <br>
    Null Hypothesis $(H_0) : \sum_1 = \sum_2$ <br>
    Alternative Hypothesis $(H_1)$ : $\sum_1 > \sum_2$
    
    **Kesimpulan** <br>
    Pertimbangkan significance level alpha adalah 5% atau 0.05. Tingkat signifikansi 5% atau kurang berarti ada kemungkinan
    95% atau lebih besar bahwa hasilnya tidak acak.<br>
    Di sini p-value lebih rendah dari significance level yang kita tentukan yaitu 0.05. Jadi, kita reject the null
    hypothesis. Artinya, terdapat perbedaan yang signifikan dalam proporsi penduduk yang memiliki resedential units di
    Bronx dan Brooklyn.
